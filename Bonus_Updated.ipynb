```python
import pandas as pd
import numpy as np

# Step 1: Format Pay Period Date
df_comp['Pay Period Start Date'] = pd.to_datetime(df_comp['Pay Period Start Date'])
df_comp['Month'] = df_comp['Pay Period Start Date'].dt.to_period('M').astype(str)

# Step 2: Filter out Bonus/Callback/Expenses
excluded = ['Bonus', 'call back', 'expenses']
df_filtered = df_comp[~df_comp['Pay Code Label'].str.lower().isin([e.lower() for e in excluded])].copy()

# Step 3: Calculate Effective Hours, if 'Pay Code Label' is 'CALL', use 'Schedule Duration (hours)', otherwise use 'Hours Rounded'
df_filtered['Effective Hours'] = np.where(
    df_filtered['Pay Code Label'].str.upper() == 'CALL',
    df_filtered['Schedule Duration (hours)'],
    df_filtered['Hours Rounded']
)

# Step 4: Location-wise group
location_hours = df_filtered.groupby(
    ['Staff First Name', 'Staff Last Name', 'Month', 'Location ID'],
    as_index=False
)['Effective Hours'].sum().rename(columns={'Effective Hours': 'Location_Hours'})

# Step 5: Monthly total hours
total_hours = location_hours.groupby(
    ['Staff First Name', 'Staff Last Name', 'Month'],
    as_index=False
)['Location_Hours'].sum().rename(columns={'Location_Hours': 'Total_Hours'})

# Step 6: Merge for percentages
result = pd.merge(location_hours, total_hours,
                  on=['Staff First Name', 'Staff Last Name', 'Month'])

result['Location_Percentage'] = round(
    (result['Location_Hours'] / result['Total_Hours']) * 100, 2
)

# Step 7: Prepare Bonus Data
df_bonus['Facility'] = df_bonus['Facility'].fillna('null')
df_bonus = df_bonus.rename(columns={'Total Amount': 'Bonus'})

# Split Name into first and last
df_bonus[['Staff First Name', 'Staff Last Name']] = df_bonus['Name'].str.extract(r'^(\w+)\s+(.*)$')

# Step 8: Merge Bonus into result
final_df = pd.merge(
    result,
    df_bonus,
    on=['Staff First Name', 'Staff Last Name'],
    how='left'
)

# Step 9: Filter based on Facility
final_df['Facility'] = final_df['Facility'].str.upper()
final_df['Location ID'] = final_df['Location ID'].str.upper()

final_df['Is_Matched_Facility'] = np.where(
    (final_df['Facility'] == 'NULL') | (final_df['Facility'] == final_df['Location ID']),
    True,
    False
)
final_df = final_df[final_df['Is_Matched_Facility']].copy()

# Step 10: Bonus Distribution
final_df['Bonus_Distributed'] = round(
    final_df['Bonus'] * final_df['Location_Percentage'] / 100, 2
)

# Final Output
final_output = final_df[[
    'Staff First Name', 'Staff Last Name', 'Month', 'Location ID',
    'Location_Hours', 'Total_Hours', 'Location_Percentage',
    'Bonus', 'Bonus_Distributed'
]]

# Display
display(final_output)
```
